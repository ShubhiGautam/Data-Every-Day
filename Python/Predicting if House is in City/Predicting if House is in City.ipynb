```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./houses_to_rent.csv')
data.head()
```

```python
data.shape
```

```python
data.drop(data.columns[0], axis=1, inplace=True)
data.head()
```

```python
data['floor'].replace(to_replace ='-', value=0, inplace=True)
data.head()
```

```python
data['animal'].replace(to_replace='not acept', value=0, inplace=True)
data['animal'].replace(to_replace='acept', value=1, inplace=True)
data.head()
```

```python
data['furniture'].replace(to_replace='not furnished', value=0, inplace=True)
data['furniture'].replace(to_replace='furnished', value=1, inplace=True)
data.head()
```

```python
for col in ['hoa', 'rent amount','property tax','fire insurance','total']:
    data[col].replace(to_replace='R\$', value='', regex=True, inplace=True)
    data[col].replace(to_replace=',', value='', regex=True, inplace=True)
data.head()
```

```python
data['hoa'].replace(to_replace='Sem info', value='0', inplace=True)
```

```python
data['hoa'].replace(to_replace='Incluso', value='0', inplace=True)
data['property tax'].replace(to_replace='Incluso', value='0', inplace=True)
```

```python
data.dtypes
```

```python
#casting the entire dataset into integers

data = data.astype(dtype=np.int64)
```

```python
data.isin(['Incluso']).any()
```

```python
data.dtypes
```

```python
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
#split data between X and Y

y = data['city']
X = data.drop('city', axis=1)
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X) #finding min max
X = scaler.transform(X) #transforming it between 0 and 1
X #numpy array
```

```python
pd.DataFrame(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
#model initalization
log_model = LogisticRegression(penalty = 'l2', verbose=1)
svm_model = SVC(kernel='rbf', verbose=1)
nn_model = MLPClassifier(hidden_layer_sizes=(16,16), activation = 'relu', solver='adam', verbose=1)
```

```python
#fit each model

log_model.fit(X_train,y_train)
svm_model.fit(X_train,y_train)
nn_model.fit(X_train,y_train)
```

```python
print(log_model.score(X_test,y_test))
print(svm_model.score(X_test,y_test))
print(nn_model.score(X_test,y_test))
```

```python
data[data.columns[0]].sum()/data.shape[0]
#data[data.columns[[0]].sum is nothing but getting the total of the 1s - how many houses are in the city, and we then take it over the total
#as we can see that 86% of our data exists within a city
#accuracy metric is only really good when you have equal number of positive and negative examples 
#this is evidently skewed data
#thus need to use a different metric -> f-score
```

```python
from sklearn.metrics import f1_score
#f1_score -> compares precision and the recall and caculates the f1score accordingly 
```

```python
log_pred = log_model.predict(X_test)
svm_pred = svm_model.predict(X_test)
nn_pred = nn_model.predict(X_test)
```

```python
log_pred
```

```python
print(f1_score(log_pred, y_test))
print(f1_score(svm_pred, y_test))
print(f1_score(nn_pred, y_test))
```

```python
#The highest possible value of an F-score is 1.0, indicating perfect precision and recall, and the lowest possible value is 0, if either the precision or the recall is zero.
```
